```python
query = (
    spark.readStream
        .format("cloudFiles")
        .option("cloudFiles.format", "csv")
        .option("cloudFiles.schemaLocation", "/Volumes/workspace/default/fraud_data/checkpoints/schema")
        .option("header", "true")
        .option("inferSchema", "true")
        .option("pathGlobFilter", "creditcard.csv")  # lock to exact filename
        .load("/Volumes/workspace/default/fraud_data/")
    .writeStream
        .format("delta")
        .trigger(availableNow=True)
        .option("checkpointLocation", "/Volumes/workspace/default/fraud_data/checkpoints/bronze")
        .outputMode("append")
        .toTable("fraud_bronze")
)

query.awaitTermination()
print("Bronze ingestion complete.")
```

```python
df_bronze = spark.table("fraud_bronze")
print(f"Row count: {df_bronze.count()}")
df_bronze.printSchema()
df_bronze.show(5)
```
